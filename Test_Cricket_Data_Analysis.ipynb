### Q-8: Remove a column from the dataframe

```python
# Remove BBI and BBM columns as they have unorganized data.

df.drop(['Best_Bowling/Innings' , 'Best_Bowling/Match'], axis=1, inplace=True)

display(df.head())
```

### Removing & Splitting Player name and Country from Player column

```python
#Removing the closing bracket

df['Player'] = df['Player'].str.replace(")"," ", regex=True)  

#splitting

df_country=df['Player'].str.split("(", expand=True)
```

```python
display(df_country.head())
```

```python
# Dropping the old player column

df.drop('Player', axis=1 ,inplace=True)  
display(df.head())
```

#### Concatinating the new Dataframe & showing the new dataset with changes

```python
# concatenating the new data set with the main data set
df = pd.concat([df, df_country], axis=1) 
#display(df.head())

# renaming new splitted column
df = df.rename(columns={0: 'Player',1: 'Country'})  
#display(df.head())

# new column sequence
column_sequence = ['Player' , 'Country' , 'Span' , 'Match' , 'Innings', 'Balls', 'Runs' , 'Wickets', 'Average' , 'Economy_Rate' , 'Strike_Rate' , '5_Wickets' , '10_Wickets']
df = df[column_sequence]


# Checking the new column sequence in the Data Set
display(df.head())
```

### Q: How many players played for ICC?

```python
# Creating a function to check if played in ICC or not 

def ICC_status(x):
    if "ICC" in x:
        return "Yes"
    else:
        return "No"

# Applying function and creating new column for playing in ICC
df['ICC_status'] = df['Country'].apply(ICC_status)

# Removing "ICC/" from the country column
df['Country'] = df['Country'].str.replace("ICC/","", regex=True)
display(df.head(5))
```

```python
# Count number of total players who played for ICC

print("Number of total ICC ICC_Playing Players: ")
print(df['ICC_status'].value_counts())
```

##### Observation

> *We see, there are only **5 ICC Playing players** in the data set, where **74 players didn't play for ICC**.*

#### Reorganizing the dataframe after splitting start and end year

```python
# splitting 'span' column
df_span = df['Span'].str.split("-", expand=True)

# concatenating the new dataframe with the main dataframe
df = pd.concat([df, df_span], axis=1)

# renaming new splitted column
df = df.rename(columns={0: "Start_Year",
                        1: "End_Year"})

# removing 'span' column
df = df.drop("Span", axis=1)

# new column sequence as a list
column_sequence = ['Player', 'Country','Start_Year', 'End_Year', 'Match' , 'Innings', 'Balls', 'Runs' , 'Wickets', 'Average' , 'Economy_Rate' , 'Strike_Rate' , '5_Wickets' , '10_Wickets', 'ICC_status']

# implementing the new column sequence
df = df[column_sequence]
```

#### Changing Start and End Year from object type to integer

```python
df['End_Year']=pd.to_numeric(df['End_Year'])
df['Start_Year']=pd.to_numeric(df['Start_Year'])
```

```python
# total years of career of the players

df['Career_Span']=df['End_Year']-df['Start_Year']

# rearranging columns
column_sequence = ['Player', 'Country','Start_Year', 'End_Year', 'Career_Span' , 'Match', 'Innings', 'Balls', 'Runs', 'Wickets' , 'Average' , 'Economy_Rate' , 'Strike_Rate' , '5_Wickets' , '10_Wickets', 'ICC_status']
df = df[column_sequence] 
display(df.head())
```
